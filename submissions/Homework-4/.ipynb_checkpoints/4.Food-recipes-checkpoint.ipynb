```python
import numpy as np
import scipy.cluster.hierarchy as hr
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
%matplotlib inline
sorted_countries = sorted(country_ingredients)

soysause_countries = []

for c in sorted_countries:
    print(c)
    if 'butter' in country_ingredients[c].keys():
        val = country_ingredients[c]['butter']
        soysause_countries.append(val)
      
    else:
        soysause_countries.append(0)
        
print soysause_countries
CN_US = [sorted_countries[0],sorted_countries[1],sorted_countries[2],sorted_countries[9]]
print sorted_countries
print CN_US
SS_C = [soysause_countries[0],soysause_countries[1],soysause_countries[2], soysause_countries[9]]
print SS_C


fig, ax = plt.subplots(figsize = (10,10))
x_coord = np.arange(len(CN_US))
width = 0.1
ax.set_xticks(x_coord+width/2)
ax.set_xticklabels(CN_US)
plt.title('Disparity Between Most Popular Ingredient in America and Other Countries of World')
plt.bar(x_coord, SS_C, color = 'r', width =width)
```

```python
sorted_ingredients = sorted(ing)
data_countries = np.zeros(shape =(len(sorted_countries), len(ing)))
for k in country_ingredients.keys():
    diction = country_ingredients[k]
    for v in diction.keys():
        i = sorted_countries.index(k)
        j = sorted_ingredients.index(v)
        data_countries[i][j] = diction[v]
print(data_countries)
        
Z2 = hr.linkage(data_countries, method='complete', metric='euclidean')
fig2 = plt.figure(figsize=(15,15))
T2 = hr.dendrogram(Z2, color_threshold = 350, leaf_font_size=16)
plt.title('Dendogram with Countries and Frequency of Ingredients')

fig2.show()
print sorted_countries
```
