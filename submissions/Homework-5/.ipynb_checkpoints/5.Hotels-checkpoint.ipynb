````python
# Task3
# using methods in lecture12


data = pd.read_csv('hotels1.csv')
data['Total'] = data['Star'] * data['TotalReview']

y = data['Total']
X = data.drop(['Total'],axis=1)
model = sm.OLS(y, X)
print model

results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params
````

````python
# Task4
data['Score'] = (data['Excellent'] / data['TotalReview']) > 0.6
print data
````

````python
y = data['Score']
X = data.drop(['Score'],axis=1)

data.SleepQuality.value_counts().sort_index().plot(kind='bar')
plt.title('Histogram of Sleep Qulity')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

data.Room.value_counts().sort_index().plot(kind='bar')
plt.title('Histogram of Room')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

data.Location.value_counts().sort_index().plot(kind='bar')
plt.title('Histogram of Location')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

data.Cleaniness.value_counts().sort_index().plot(kind='bar')
plt.title('Histogram of Cleanliness')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

data.Value.value_counts().sort_index().plot(kind='bar')
plt.title('Histogram of Value')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()


data.Service.value_counts().sort_index().plot(kind='bar')
plt.title('Histogram of Service')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
````

````python
# evaluate the model by splitting into train and test sets

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,
                                                                     y,
                                                                     test_size=0.2,
                                                                     random_state=0
                                                                     )
logistic_regr = linear_model.LogisticRegression()
logistic_regr.fit(X_train, y_train)
````

````python

y_predicted = logistic_regr.predict(X_test)
print y_predicted

probs = logistic_regr.predict_proba(X_test)
print probs
print " "

print metrics.accuracy_score(y_test, y_predicted)
print " " 
print metrics.confusion_matrix(y_test, y_predicted)
print metrics.classification_report(y_test, y_predicted)

print " "
print pd.DataFrame(zip(X.columns, np.transpose(logistic_regr.coef_)))

# evaluate the model using 10-fold cross-validation
scores = cross_validation.cross_val_score(linear_model.LogisticRegression(),
                                          X,
                                          y,
                                          scoring='accuracy',
                                          cv=10)
print scores
print scores.mean()
````
